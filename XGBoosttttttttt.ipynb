```
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Colejoby/DW/master/DIPEP.csv', header=None)
df.head()
```

```
df_clean=df.dropna()
df_clean.dtypes
```

```
df.isna().any()
```

#แยกเลเบล

```
y=df_clean.iloc[:,0]
X=df_clean.iloc[:,1:401]
```

```
y.shape
```

```
X.shape
```

#แยก test train 

```
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y,test_size=0.25 ,random_state=0)
```

```
X_train.shape
```

```
X_val.shape
```

#Normlization

```
from sklearn.preprocessing import MinMaxScaler
scl = MinMaxScaler()
scl.fit(X_train)
X_train_norm = scl.transform(X_train)
X_val_norm = scl.transform(X_val)
```

#LogisticRegression

```
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(C=32,intercept_scaling=0.5)
clf
```

```
clf.fit(X_train_norm,y_train)
```

```
y_predict = clf.predict(X_val_norm)
y_predict
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
clf = LogisticRegression(C=32,intercept_scaling=0.5)
accuracy = cross_val_score(clf, X_train_norm, y_train, cv=5)
accuracy.mean()
```

```
clf.fit(X_train_norm,y_train)
y_predict = clf.predict(X_val_norm)
sum(y_predict==y_val) / len(y_val)
```
